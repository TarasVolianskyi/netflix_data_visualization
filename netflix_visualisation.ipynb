```python
df2=pd.pivot_table(df, index=["release_year"], values='count', aggfunc="count")
df2
```

```python
years = list(df2.index)
values = list(df2['count'])
  
fig = plt.figure(figsize = (15, 5))
 
plt.bar(years, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("release_year")
plt.ylabel("count")
plt.show()
```

```python
df4=pd.pivot_table(df[df.release_year>=2000], index=["release_year"], columns=['type'], values='count', aggfunc="count")
```

```python
import matplotlib.pyplot as plt


labels = list(df4.index)


fist_indicator = list(df4['Movie'])
secoond_indicator = list(df4['TV Show'])

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig = plt.figure(figsize = (15, 5))

ax.bar(labels, fist_indicator, width, label='Movie')
ax.bar(labels, secoond_indicator, width,  bottom=fist_indicator,
       label='TV Show')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
